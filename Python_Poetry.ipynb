### 1. Creating the poetry project

```python
# creating the scalable_machine_learning directory on Drive
!mkdir scalable_machine_learning
%cd scalable_machine_learning
```

```python
# installing poetry using pip
!pip install poetry
!poetry config warnings.export false
```

```python
!poetry --version
```

### 2. Creating the  virtual environment

```python
# configuring poetry to create virtual environments in the project folder
!poetry config virtualenvs.in-project true
```

```python
# creating the pyproject.toml file. Since it is a new project,
# we need to initialize it with "poetry". So, running "poetry init" and but not specifying any dependencies for this phase.
!poetry init
```

### 3. Adding the packages

```python
'''
    Since we ran this command for the first time, a virtual environment is created in the project folder. By default, it is called *.venv*
    The versions for packages are indicated
    The "poetry.lock" file is created

    pyspark mlflow kafka-python kafka numpy pandas scikit-learn PyYAML python-box ensure tqdm types-PyYAML Flask Flask-Cors kubernetes SQLAlchemy joblib boto3 matplotlib seaborn jupyter pytest hdfs gensim workalendar pathlib xgboost

'''
!poetry add pyspark mlflow kafka-python kafka numpy pandas scikit-learn PyYAML python-box ensure tqdm types-PyYAML Flask Flask-Cors kubernetes SQLAlchemy joblib boto3 matplotlib seaborn jupyter pytest hdfs gensim workalendar pathlib xgboost
```
